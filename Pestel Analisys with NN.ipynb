```
import numpy as np
a=np.matrix([[1.00,0.25,0.20,0.20,0.33,3.00],
             [4.00,1.00,1.00,1.00,2.00,1.00],
             [5.00,1.00,1.00,1.00,4.00,2.00],
             [5.00,1.00,1.00,1.00,2.00,3.00],
             [3.00,0.50,0.25,0.50,1.00,3.00],
             [0.33,1.00,0,0.33,0.33,1.00]])
print (a)
```

```
#print (a.sum(axis=0)) #Suma por columnas
b=a/a.sum(axis=0) # Se normalizan la matriz

eigen=1.0/6.0*(b.sum(axis=1))# se encuentra el promedio
print(eigen)
```

```
# Fuzzy Cognitive Map
E=np.matrix([[0,0.8,0.8,0.8,0.5,0.2],[0,0.8,0,0.5,0.4,0.2],
             [0.8,0.8,0,0.5,0.5,0.4],
             [0.8,0.5,0.5,0,0.4,0.4],
             [0.5,0.4,0.5,0.4,0,0.2],
             [0.2,0.2,0.4,0.4,0.2,0]])
print(E)
```

```
def steady(M):    
    C=M
    I=np.identity(len(M))
    M.dot(I)
    flag=True
    print(flag)
    c=0
    C=M
    T=M
    while flag:               
        c=c+1
        #print(flag)
        C=M.dot(C+I)       
        #print(T)
        #print(C)        
        flag=not((C==T).all())
        #print(flag)
        if c==50:
            flag=False
        T=C
    #print(c)
    return C
Ef=steady(E)   
print(Ef)
```

```
eigenn=(eigen/np.amax(eigen))
#print(eigenn)
# Normalizar la matriz

Efn=Ef/np.amax(Ef.sum(axis=1)) #row sum

w=eigenn+Efn.dot(eigenn)

#print(w)
#Finally, normalize the global weight (W).
print(w/w.sum(axis=0))
```
